```python
def user_initialize():
    print("Hello! Welcome to TictacToe!")
    
    acceptchar = ['X','O']
    
    global p1character
    p1character = 'intlwrong'
    
    while p1character not in acceptchar:
        
        p1character = input('Player 1, choose your character (X or O): ')
        
        if p1character not in acceptchar:
            clear_output()
            print('Invalid selection. Please select either X or O.')
            
        else:
            pass
    
    global p2character
    if p1character == 'X':
        p2character = 'O'
        
    elif p1character == 'O':
        p2character = 'X'
        
    clear_output()
    
    print("Player 1 : {}".format(p1character))
    print("Player 2 : {}".format(p2character))
```

```python
def check_win():
    
    global player1win

#check vertical wins

    if row1[0] == p1character and row2[0] == p1character and row3[0] == p1character:
        player1win = True
    elif row1[1] == p1character and row2[1] == p1character and row3[1] == p1character:
        player1win = True
    elif row1[2] == p1character and row2[2] == p1character and row3[2] == p1character:
        player1win = True
    else:
        player1win = False

#check horizontal wins
    
    if row1 == [p1character, p1character, p1character]:
        player1win = True
    elif row2 == [p1character, p1character, p1character]:
        player1win = True
    elif row3 == [p1character, p1character, p1character]:
        player1win = True
    
#check diagonal wins
    
    if row1[0] == p1character and row2[1] == p1character and row3[2] == p1character:
        player1win = True
    elif row1[2] == p1character and row2[1] == p1character and row3[0] == p1character:
        player1win = True
    else:
        False


    global player2win

#check vertical wins

    if row1[0] == p2character and row2[0] == p2character and row3[0] == p2character:
        player2win = True
    elif row1[1] == p2character and row2[1] == p2character and row3[1] == p2character:
        player2win = True
    elif row1[2] == p2character and row2[2] == p2character and row3[2] == p2character:
        player2win = True
    else:
        player2win = False

#check horizontal wins
    
    if row1 == [p2character, p2character, p2character]:
        player2win = True
    elif row2 == [p2character, p2character, p2character]:
        player2win = True
    elif row3 == [p2character, p2character, p2character]:
        player2win = True
    
#check diagonal wins
    
    if row1[0] == p2character and row2[1] == p2character and row3[2] == p2character:
        player2win = True
    elif row1[2] == p2character and row2[1] == p2character and row3[0] == p2character:
        player2win = True
    else:
        False
        
    if player1win == True:
        print ('PLAYER 1 WINS')
    elif player2win == True:
        print('PLAYER 2 WINS')
    elif row1[0] and row1[1] and row1[2] and row2[0] and row2[1] and row2[2] and row3[0] and row3[1] and row3[2] == 'X' or 'O' and player1win and player2win == False:
        print('TIE!')
```

```python
def playfinalgame():

#initialize
    user_initialize()
    global possiblespots
    possiblespots= ['1','2','3','4','5','6','7','8','9']
    
#display board
    display()

#set turn
    global turn
    turn = 'player1'
    
    player1win = False
    player2win = False

    while player1win == False and player2win == False:
            
        #accept player 1 input
        if turn == 'player1':
    
            p1_input()

        #accept player 2 input
        elif turn == 'player2':
    
            p2_input()

        #return updated board
        clear_output()
        print("Player 1 : {}".format(p1character)) 
        print("Player 2 : {}".format(p2character))
        update_board()
            
    #check for win and tie
        check_win()
        
     #switch turn
        if turn == 'player1':
            turn = 'player2'
        elif turn == 'player2':
            turn = 'player1'
               
#play again prompt
    if input("Would you like to play again? True/False") == True:
        playfinalgame()
    else:
        print ('Thank you for playing!')
    
```
